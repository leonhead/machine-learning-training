```python
# 1) import the data
import pandas as pd

train_df = pd.read_csv('train.csv',delimiter=',')
test_df = pd.read_csv('test.csv',delimiter=',')

combine = [train_df,test_df]
```

```python
train_df.head(10)
```

```python
#2) Cleanup
def print_train_and_test_missings():
    print("Traindata:\n%s" % train_df.isnull().sum())
    print("-"*30)
    print("Testdata:\n%s" % test_df.isnull().sum())
    
print_train_and_test_missings()
```

```python
freq_port = train_df.Embarked.dropna().mode()[0]

combine = [train_df, test_df]

for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_port)
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch']+1
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] == 1, 'IsAlone'] = 1
    
train_df = train_df.drop('Cabin',axis=1).copy()
test_df = test_df.drop('Cabin',axis=1).copy()

train_df = train_df.drop('Ticket',axis=1).copy()
test_df = test_df.drop('Ticket',axis=1).copy()

train_df = train_df.drop('FamilySize',axis=1).copy()
test_df = test_df.drop('FamilySize',axis=1).copy()

train_df = train_df.drop('SibSp',axis=1).copy()
test_df = test_df.drop('SibSp',axis=1).copy()

train_df = train_df.drop('Parch',axis=1).copy()
test_df = test_df.drop('Parch',axis=1).copy()

train_df = train_df.drop('Name',axis=1).copy()
test_df = test_df.drop('Name',axis=1).copy()

train_df = train_df.drop('PassengerId',axis=1).copy()
test_df = test_df.drop('PassengerId',axis=1).copy()


test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].mean())
    
combine = [train_df, test_df]
```

```python
train_df = train_df.dropna()
test_df = test_df.dropna()
    
combine = [train_df, test_df]
```

```python
print_train_and_test_missings()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for dataset in combine:
    dataset['Embarked'] = le.fit_transform(dataset['Embarked'])

for dataset in combine:
    dataset['Sex'] = le.fit_transform(dataset['Sex'])   
    
```

```python
train_df['AgeBand'] = pd.qcut(train_df['Age'], 4)
train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
train_df['FareBand'] = pd.qcut(train_df['Fare'], 4)
train_df[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

```python
for dataset in combine:
    dataset.loc[ dataset['Age'] <= 20.125, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 20.125) & (dataset['Age'] <= 28.0), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 28.0) & (dataset['Age'] <= 38.0), 'Age']   = 2
    dataset.loc[ dataset['Age'] > 38.0, 'Age'] = 3
    dataset['Age'] = dataset['Age'].astype(int)
    
train_df = train_df.drop(['AgeBand'], axis=1)
combine = [train_df, test_df]
```

```python
for dataset in combine:
    dataset.loc[ dataset['Fare'] <=  8.05, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 8.05) & (dataset['Fare'] <= 15.742), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 15.742) & (dataset['Fare'] <=  33.375), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] >  33.375, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
    
train_df = train_df.drop(['FareBand'], axis=1)
combine = [train_df, test_df]
```
