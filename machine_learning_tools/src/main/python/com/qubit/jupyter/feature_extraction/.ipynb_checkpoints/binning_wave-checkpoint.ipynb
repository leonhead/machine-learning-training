```python
import mglearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline

X, y = mglearn.datasets.make_wave(n_samples=100) 
line = np.linspace(-3, 3, 1000, endpoint=False).reshape(-1,1)

reg = LinearRegression().fit(X,y)

plt.plot(line, reg.predict(line), label="lineare Regression",c='r')
plt.plot(X[:,0], y, 'o',c='b')
plt.legend(loc="best")
```

```python
bins = np.linspace(-3,3,11)
print("bins: {}".format(bins))
```

```python
which_bins = np.digitize(X, bins=bins)
print("Datanpunkte: \n", X[:10])
print("Klassenzugehörigkeit der Datenpunte:\n", which_bins[:10])
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoder.fit(which_bins)
X_binned = encoder.transform(which_bins)
print(X_binned[:10])
```

```python
X_binned.shape
```

```python
line_binned = encoder.transform(np.digitize(line, bins=bins))
reg = LinearRegression().fit(X_binned, y)
plt.plot(line, reg.predict(line_binned), label="linear Regression mit Binning",c='r')
plt.plot(X[:,0], y, 'o',c='b')
plt.legend(loc="best")
```

```python
X_combined = np.hstack([X, X_binned])
print(X_combined.shape)
```

```python
reg = LinearRegression().fit(X_combined, y)
line_combined = np.hstack([line, line_binned])
plt.plot(line, reg.predict(line_combined), label='kombinierte lineare Regression', c='r')

for bin in bins:
    plt.plot([bin,bin], [-3,3], ':', c='k')
  
plt.plot(X[:,0],y,'o', c='b')
plt.legend(loc="best")
```

```python
X_product = np.hstack([X_binned, X * X_binned])
print(X_product.shape)
```

```python
reg = LinearRegression().fit(X_product, y)

line_product = np.hstack([line_binned, line * line_binned])
plt.plot(line, reg.predict(line_product), label='kombinierte lineare Regression', c='r')

for bin in bins:
    plt.plot([bin,bin], [-3,3], ':', c='k')
  
plt.plot(X[:,0],y,'o', c='b')
plt.legend(loc="best")
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=10, include_bias=False)
poly.fit(X)
X_poly = poly.transform(X)
print("X_poly.shape: {}".format(X_poly.shape))
```

```python
print("Einträge in X:\n{}".format(X[:5]))
print("Einträge in X_poly:{}".format(X_poly[:5]))
```

```python
print("Name der polynomiellen Merkmale:\n{}".format(poly.get_feature_names()))
```

```python
reg = LinearRegression().fit(X_poly, y)
line_poly = poly.transform(line)
plt.plot(line, reg.predict(line_poly), label='polynomielle lineare Regression', c='r')

for bin in bins:
    plt.plot([bin,bin], [-3,3], ':', c='k')
  
plt.plot(X[:,0],y,'o', c='b')
plt.legend(loc="best")
```

```python
from sklearn.svm import SVR

for gamma in [1,10]:
    svr = SVR(gamma=gamma).fit(X,y)
    plt.plot(line, svr.predict(line), label="SVR gamma={}".format(gamma))

plt.plot(X[:,0],y,'o', c='b')
plt.legend(loc="best")
```
