```python
%matplotlib inline
import os
import pandas as pd

BITCOIN_PATH = "/home/ml/ml"
CSV_FILE_NAME= "dump.csv"

def load_bitcoin_data(bitcoin_path=BITCOIN_PATH, csv_file=CSV_FILE_NAME):
    csv_path = os.path.join(bitcoin_path, csv_file)
    return pd.read_csv(csv_path, sep=';')

bitcoin = load_bitcoin_data()
bitcoin.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def looking_correlations():
    corr_matrix = bitcoin.corr()
    corr_matrix["PRICE_EUR"].sort_values(ascending=False)
    f, ax = plt.subplots(figsize=(11,9))
    sns.heatmap(corr_matrix, annot=True)
    
looking_correlations()
```
