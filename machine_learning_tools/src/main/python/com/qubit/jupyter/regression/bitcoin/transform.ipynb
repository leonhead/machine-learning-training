```python
%matplotlib inline
import os
import pandas as pd

PROJECT_PATH = "/home/ml/project/bitcoin"
CSV_FILE_NAME= "data/order.csv"

def load_csv_data(project_path=PROJECT_PATH, csv_file=CSV_FILE_NAME):
    csv_path = os.path.join(project_path, csv_file)
    return pd.read_csv(csv_path, sep='\t')



data = load_csv_data()
data.tail()
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

def normilaze_transform(data=data):
    encoder = LabelEncoder()
    symbol_cat = data["SYMBOL"]
    symbol_encoded = encoder.fit_transform(symbol_cat)
    return symbol_encoded

symbol_encoded = normilaze_transform()

data_encoded_values = data.values
symbol_index = data.columns.get_loc("SYMBOL")
data_encoded_values[:,symbol_index] = symbol_encoded
print data_encoded_values
data_transform = pd.DataFrame(data=data_encoded_values,columns=data.columns.values)
data_transform.set_index('ID', inplace=True)
data_transform.head()

```

```python
def save_csv_data(project_path=PROJECT_PATH, csv_file=CSV_FILE_NAME, data_transform=data_transform):
    csv_path = os.path.join(project_path, csv_file)
    data_transform.to_csv(csv_path)

save_csv_data(PROJECT_PATH,"data/order_transformed.csv",data_transform)
```
